### Example 3

##### 1) From `left to right` : The indexing starts from <font color = 'red'>0</font> and the last element is <font color='red'>N-1</font>, where N is the length of the list.  
##### 2) From `right to left` : The indexing starts from <font color = 'red'>-1</font> and the last element is <font color = 'red'>-N</font>, where N is the length of the lsit.  


```python
# Left to Right 
print(lst[0])               # First Element
print(lst[len(lst)-1])      # Last Element

# Right to left 
print(lst[-1])              # First Element
print(lst[-len(lst)])       # Last Element
```
