### Example 5

Elements of a tuple are immutable, but if an element referene to an object that is mutable, than we can edit the state of that object. 



```python
tup = ([1,2,3], 'apple', 'mango')
print (type(tup))

# Elements of a tuple are immutable, but if an element referene to an object that is mutable, than we can edit the state of that object. 
tup[0][0] = 'Banana'

print(tup)
```

### Example 6

Tuple to List and vise-versa

```python
# List to Tuple
lst = [10,20,30]
tup = tuple(lst)
print(type(tup), tup)

# Tuple to List
tup = ('abc', 'def','ghi')
lst = list(tup)
print(type(lst), lst)

# once tuple is converted to list we can edit as per our need. 
lst[0] = 100
print(lst)
```
