### Example 4

However, we can define that `__add__` method in our custom object, to tell Python how to add two vectors together:

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __repr__(self):
        return f'Vector({self.x}, {self.y})'
```

```python
v1 = Vector(1, 1)
v2 = Vector(2, 3)
print(v1)
print(v2)
```

```python
v1 + v2         # NOTE : At this point python does not know to add vectors tuples. 
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        
    def __repr__(self):
        return f'Vector({self.x}, {self.y})'
```

```python
v1 = Vector(1, 1)
v2 = Vector(2, 3)
print(v1)
print(v2)
```

```python
v1 + v2
```
