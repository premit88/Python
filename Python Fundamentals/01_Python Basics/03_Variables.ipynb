## 3) Variables

1) Think object as an box, which as state (data/value) and functionality asociated with it. Variable is just a label that is stuck on that box. But at any point in time we can rip off that label and stick on another box. 

2) Variables names are case sensitive 







<font color = 'red'> **Remember,  State = data and Method = function** </font>












### Example 1

```python
a = 10              # Left side of the assignment operator should be a label (Variable)
print(a)
b = a + 11
print(b)

a = 3.14            # Variables can be reassigned. 
print(a)
```

### Example 3

```python
a = float(10)
a
```

```python
float = 100.5     # NOTE : Do not use reserve words as variables, else you wont be able to use them later on. 
print(float)
```

```python
a = float(10)   # At this point, now float is a variable and not a function. 
```

```python
del float       # Delete the old variable and then you would be able to use it. 
b = float(10)
print(b)
```
