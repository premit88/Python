## 7) Comparison Opertors















### Example 1
1) We saw that we have two types of equality comparison:
- value comparison (`==`)
- object identity (`is`)
2) When we use the `==` operator, the type itself (an object with functionality and state), specifies how two different objects may be equal, in some sense.
3) For numeric types, the `==` comparison compares the **value** of the number.


```python
x = 10          # 10 is an int object with state(value) as 10. 'x' is just a label on the object. 
y = 10.0        # 10.0 is an float object with state(value) as 10.0. 'y' is just a label on the object. 
print ( x == y)
print ( x is y)
```

### Example 2

One way we can investigate objects and whether two objects are the same objects, is to look at their **memory address**.
Objects in Python reside somewhere in memory. Think of memory as a bunch of slots, each of which has an **address** - just like mailboxes have addresses.
If the memory address of an object is the same as the other one, then they are the same object - just like two addresses that are the same correspond to the same mailbox.
If two objects are identical (the same object), they are obviously equal too (two mailboxes that correspond to the same address **are** the same mailbox, so they must have the same content).
In Python we can use the `id` function to see the memory address of an object:


```python
a = 10
b = 10.0
print ( a == b)
print (a is b)
print (id(a), id(b))
```

### Example 3 - Trick
For small numbers python pre creates object. So it is advisable to avoid is comparing object identity and just use equality comparison

```python
# Small number
a = 10
b = 10
print ( a is b)
print (id(a), id(b))

# Big number
a = 10_000
b = 10_000
print ( a is b)
print (id(a), id(b))
```
