## 6) Integer Divison and Modulus 











### Example 1

```python
print (10 / 3)
print (10 // 3)
print (-10 / 3)
print (-10 // 3)
print(10 % 3)

```

### Example 2

```python
print (10 % 2)      # Even number check
print (5 % 2)       # Odd number check
```

### Example 3

```python
elapsed_minutes = 623
hours = elapsed_minutes // 60
minutes = elapsed_minutes % 60
print(hours, minutes)
```
