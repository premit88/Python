## 2) Objects

















### Example 1
1) Below we create a custom object called Account, which has states(data/value) like account number, account type and balance and it also has funcationality like deposit and withdraw. 
2) NOTE: Everything is python is a object, when an object is defined as Class (blueprint), it can be use to create more objects. 

```python
class Account:
    def __init__(self, account_number, account_type, initial_balance):
        self.account_number = account_number    
        self.account_type = account_type
        self.balance = initial_balance
        
    def deposit(self, amount):
        # should also check that amount is a numerical value!
        if amount > 0:
            self.balance = self.balance + amount
            print(f'Deposited {amount}')
            print(f'New balance is: {self.balance}')
        else:
            print(f'{amount} is an invalid amount.')
            
    def withdraw(self, amount):
        # should also check that amount is a numerical value!
        if amount > 0 and amount <= self.balance:
            self.balance = self.balance - amount
            print(f'Withdrawal: {amount}')
            print(f'New Balance: {self.balance}')
        else:
            if amount < 0:
                print(f'{amount} is an invalid amount')
            else:
                print('Insufficient funds.')
                print(f'Current balance is {self.balance}')
```

```python
my_account = Account('123-456', 'savings', 1000)    # NOTE: my_account is new object created using blueprint(class) named Account. 
```

```python
my_account.account_number       # account number is a state of the object my_account. 
```

```python
my_account.account_type         # account type is a state of the object my_account.
```

```python
my_account.balance              # account balance is a state of the object my_account.
```

```python
my_account.deposit(1000)        # NOTE : To use functionality of an objet it needs to be called by using () brackets. Deposit functionlaity called of the objet my_account.  
```

```python
my_account.withdraw(500)        # Withdraw functionlaity called of the objet my_account. 
```
